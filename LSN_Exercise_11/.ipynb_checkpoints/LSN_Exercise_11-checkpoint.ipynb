## Exercise 11.1
Valuto il comportamento della regressione lineare effettuata tramite la rete neurale variando il numero di epoche, il numero di valori di training e il rumore dei dati di validazione. Per questo esercizio utilizzo una rete neurale semplice composta da un singolo neurone come nell'esempio riportato nel *notebook* relativo all'undicesima esercitazione.

### <font color='blue'> Epochs </font> 
Valuto il comportamento della regressione lineare variando il numero di epoche ovvero il numero di iterazioni sull'intero insieme di dati in input. Per ogni grafico ho utilizzato $100$ valori di training e $15$ dati di validazione in assenza di rumore.
I grafici ottenuti mostrano che la predizione della rete neurale diventa precisa effettuando molte iterazioni sui dati di input.

```python
import math
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

def target(x):
    return 2*x + 1

plt.figure(figsize=(20, 12))
sub = [231, 232, 233, 234, 235, 236]
NN = ['10', '30', '50', '75', '100', '150']

sigma = 0
np.random.seed(0)
x_valid = np.linspace(-1, 1, 15)
y_valid = np.random.normal( target(x_valid), sigma )
x_predicted = np.linspace(-1, 1, 100)

for i in range(6):
    savedpath = "Saved/exercise1/modelNep" + str(NN[i])
    model = tf.keras.models.load_model(filepath=savedpath)
    y_predicted = model.predict(x_predicted).reshape(-1)

    plt.subplot(sub[i])
    title = "$N_{epochs} = " + str(NN[i]) + "$"
    plt.title(title)
    plt.scatter(x_valid, y_valid, color='r', label='validation data') #dati per validazione    
    plt.scatter(x_predicted, y_predicted, color='tab:orange', label='prediction') #predizione NN
    plt.plot(x_predicted, target(x_predicted), label='target', lw=2.5) #funzione target
    plt.grid(True)
    plt.legend()
    
plt.show()
```

### <font color='blue'> Training data </font> 
Valuto il comportamento della regressione lineare variando il numero di dati di training. Per ogni grafico ho utilizzato $30$ epoche e $15$ dati di validazione in assenza di rumore.
I grafici ottenuti mostrano che la predizione della rete neurale diventa precisa utilizzando un insieme grande di dati di training in input.

```python
def target(x):
    return 2*x + 1

plt.figure(figsize=(20, 12))
sub = [231, 232, 233, 234, 235, 236]
NN = [25, 50, 100, 250, 500, 1000]

sigma = 0
np.random.seed(0)
x_valid = np.linspace(-1, 1, 15)
y_valid = np.random.normal( target(x_valid), sigma )
x_predicted = np.linspace(-1, 1, 100)

for i in range(6):
    savedpath = "Saved/exercise1/modelNtrain" + str(NN[i])
    model = tf.keras.models.load_model(filepath=savedpath)
    y_predicted = model.predict(x_predicted).reshape(-1)

    plt.subplot(sub[i])
    title = "$N_{train} = " + str(NN[i]) + "$"
    plt.title(title)
    plt.scatter(x_valid, y_valid, color='r', label='validation data') #dati per validazione    
    plt.scatter(x_predicted, y_predicted, color='tab:orange', label='prediction') #predizione NN
    plt.plot(x_predicted, target(x_predicted), label='target', lw=2.5) #funzione target
    plt.grid(True)
    plt.legend()
    
plt.show()
```

### <font color='blue'> Rumore </font> 
Valuto il comportamento della regressione lineare introducendo del rumore nei dati di validazione.
Per ogni grafico ho utilizzato $100$ epoche, $1000$ dati di training e $15$ dati di validazione.
I grafici ottenuti mostrano che, anche in presenza di rumore, utilizzando un numero elevato di dati e di iterazioni la predizione della rete neurale rimane precisa.

```python
def target(x):
    return 2*x + 1

NN = [0.25, 0.50, 0.75]
sub = [131, 132, 133]
plt.figure(figsize=(16, 6))

np.random.seed(0)
x_valid = np.linspace(-1, 1, 15)
x_predicted = np.linspace(-1, 1, 100)

for i in range(3):
    sigma = NN[i]
    y_valid = np.random.normal( target(x_valid), sigma )
    savedpath = "Saved/exercise1/modelsigma" + str(sigma)
    model = tf.keras.models.load_model(filepath=savedpath)
    y_predicted = model.predict(x_predicted).reshape(-1)

    plt.subplot(sub[i])
    title = "$\sigma = " + str(NN[i]) + "$"
    plt.title(title)
    plt.scatter(x_valid, y_valid, color='r', label='validation data') #dati per validazione    
    plt.scatter(x_predicted, y_predicted, color='tab:orange', label='prediction') #predizione NN
    plt.plot(x_predicted, target(x_predicted), label='target', lw=2.5) #funzione target
    plt.grid(True)
    plt.legend()
    
plt.show()
```

### <font color='blue'> Neural Networks </font> 
I seguenti grafici mostrano la regressione lineare prodotta tramite la rete neurale: ogni rete  formata da tre strati e variando il numero di neuroni di ogni strato  possibile ottenere una regressione compatibilie con il target. Ogni rete neurale  composta da tre strati sequenziali densamente connessi in cui ogni neurone  connesso a tutti i neuroni dello strato precedente:

- *input layer*, composto da neuroni con attivazione *relu*
- *hidden layer*, composto da neuroni con attivazione *relu*
- *output lauer*, composto da un singolo neurone di output

Il codice qui mostrato  quello utilizzato per definire la rete neurale dell'ultimo grafico con 100 neuroni nel layer di input e 25 in quello intermedio.
<p style="border:2px; border-style:solid; border-color:#F5F5F5; padding: 1em; background-color:#F5F5F5">
<font face="Courier">model = Sequential()</font><br />
<font face="Courier">model.add(Dense(units=100, input_dim=1, activation='relu')) </font><br />
<font face="Courier">model.add(Dense(units=25, activation='relu')) </font><br />
<font face="Courier">model.add(Dense(1))</font>
</p>

Come facilmente prevedibile, il risultato migliore si ottiene per l'ultima rete neurale testata che contiene il maggior numero di neuroni. La regressione risulta corretta solo nell'intervallo $[-1,1]$ in cui si hanno i valori di training. 

```python
import math
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

def target(x):
    return 4 - 3*x - 2*x**2 + 3*x**3

plt.figure(figsize=(16, 12))
sub = [231, 232, 233, 234, 235, 236]
NN = ['NN1', 'NN2', 'NN3', 'NN4', 'NN5', 'NN6']

sigma = 0.2
np.random.seed(0)
x_valid = np.linspace(-1, 1, 15)
y_valid = np.random.normal( target(x_valid), sigma )
x_predicted = np.linspace(-1.5, 1.5, 100)

for i in range(6):
    savedpath = "Code/exercise2/model" + str(NN[i])
    model = tf.keras.models.load_model(filepath=savedpath)
    y_predicted = model.predict(x_predicted).reshape(-1)

    plt.subplot(sub[i])
    title = "$" + str(NN[i]) + "$"
    plt.title(title)
    plt.scatter(x_valid, y_valid, color='r', label='validation data') #dati per validazione    
    plt.scatter(x_predicted, y_predicted, color='tab:orange', label='prediction') #predizione NN
    plt.plot(x_predicted, target(x_predicted), label='target', lw=2.5) #funzione target
    plt.grid(True)
    plt.legend()
    
plt.show()
```

## Exercise 11.3
Sviluppo una rete neurale per effettuare la regressione della funzione bidimensionale $f(x,y) = \sin(x^2+y^2) \,,$ nell'intervallo $x,y \in [-3/2,3/2]$. Costruisco la rete neurale come per il caso precedente con tre strati sequenziali: 

- *input layer*, composto da 200 neuroni
- *hidden layer*, composto 45
- *output lauer*, composto da un singolo neurone di output

Tutti gli strati sono densamente connessi perci ogni neurone  connesso a tutti i neuroni dello strato precedente.

<p style="border:2px; border-style:solid; border-color:#F5F5F5; padding: 1em; background-color:#F5F5F5">
<font face="Courier">model = Sequential()</font><br />
<font face="Courier">model.add(Dense(units=200, input_dim=2, activation='relu')) </font><br />
<font face="Courier">model.add(Dense(units=45, activation='relu')) </font><br />
<font face="Courier">model.add(Dense(1))</font>
</p>

Il seguente grafico mostra i risultati ottenuti per tale fit utilizzando $150$ epoche, $1000$ valori di training e $100$ valori di validazione.

```python
import math
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  

a = 1.5
n = 20000
model = tf.keras.models.load_model("Saved/exercise3/model2D")

x_predicted = np.zeros((n, 2))
for i in range(n):
    x_predicted[i,0] = np.random.uniform(-a, a)
    x_predicted[i,1] = np.random.uniform(-a, a)
y_predicted = model.predict(x_predicted).reshape(-1)

x = x_predicted[:,0]
y = x_predicted[:,1]
z = y_predicted[:]

fig=plt.figure(figsize=(12, 12))
ax = fig.gca(projection='3d')

ax.scatter(x, y, z, c=z, s=5) #predizione della rete neurale
ax.set_title('Regressione neural network')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(20, 30)
plt.show()
```
