Stima del profitto delle *European option* attraverso una simulazione Monte Carlo del prezzo $S(T)$ degli asset considerati. Il profitto per i due tipi di opzioni possibili è calcolabile come:

- **Call-option** $ \quad C\left[ S(t),t \right] = \langle e^{(-rT)}max(0, S(T)-K) \rangle $
- **Put-option**$ \quad P\left[ S(t),t \right] = \langle e^{(-rT)}max(0, K-S(T)) \rangle \$

dove $r$ è l'interesse privo di rischio, $K$ il prezzo di mercato e $T$ il tempo finale della simulazione. La stima del profitto è ottenuta tramite una media sui diversi valori simulati per il prezzo $S(T)$ degli asset considerati.

### <font color='blue'> Simulazione diretta</font>

Stimo il prezzo finale degli asset con una simulazione diretta, tramite un moto Browniano geometrico $S\sim GBM(\mu,\sigma^2)$. Se il valore iniziale del prezzo è $S(0)$, utilizzo l'equazione seguente per la stima del prezzo al tempo $t$:
     $$S(t) = S(0) \exp\left[ \left(r-\frac{1}{2}\sigma^2\right)t+\sigma W(t)\right] \,,$$
     
dove $W(t)\sim\mathcal{N}(0,t)$ è un numero casuale distribuito normalmente intorno al valore $0$ e con deviazione standard $t$ e $\sigma$ è la volatilità del mercato.

Il grafico sottostante mostra il profitto ottenuto per le due opzioni al variare del numero di stime del prezzo finale e determinando le medie e le incertezze statistiche utilizzando il metodo delle **medie a blocchi**. In particolare la simulazione è stata suddivisa in $100$ blocchi, ciascuno dei quali costituito da $10^4$ stime del prezzo finale $S(T)$.

```python
import math
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 5))

#risultati analitici 
call = 14.9758
put = 5.4595

x = np.linspace(0,10**6,100)
C = call*x**0
P = put*x**0

Call, Cerr, Put, Perr= np.loadtxt('./Files/direct.out', delimiter = '  ', unpack = 'True')

plt.subplot(121)
plt.plot(x, C, color='red', label='risultato analitico')  #grafico risultato analitico
plt.errorbar(x, Call, yerr=Cerr, label='simulazione diretta') #grafico simulazione
plt.title("Call-option $C[S(T),T]$")
plt.xlabel('#stime del prezzo')
plt.ylabel('Profitto')
plt.legend()
plt.grid(True)

plt.subplot(122)
plt.plot(x, P, color='red', label='risultato analitico')  #grafico risultato analitico
plt.errorbar(x, Put, yerr=Perr, label='simulazione diretta') #grafico simulazione
plt.title("Put-option $P[S(T),T]$")
plt.xlabel('#stime del prezzo')
plt.ylabel('Profitto')
plt.legend()
plt.grid(True)

plt.show()
```

### <font color='blue'> Simulazione discretizzata</font>

Stimo il prezzo finale $S(T)$ suddividendo l'intervallo temporale $[0,T]$ in $100$ sottointervalli e valutando il prezzo in modo iterativo. In particolare se indico ciascun istante come $t_i$ per $i=1,2,..,100$ dove $t_100 = T \,,$ posso determinare il prezzo degli asset come:
    $$S(t_{i+1}) = S(t_i) \exp\left[ \left(r-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)+\sigma Z\sqrt{t_{i+1}-t_i}\right] \,,$$
dove $Z \sim \mathcal{N}(0,1)$ è un numero casuale con distribuzione di probabilità normale standardizzata.

Il grafico sottostante mostra il profitto ottenuto per le due opzioni al variare del numero di stime del prezzo finale e determinando le medie e le incertezze statistiche utilizzando il metodo delle **medie a blocchi**. In particolare la simulazione è stata suddivisa in $100$ blocchi, ciascuno dei quali costituito da $10^4$ stime del prezzo finale $S(T)$.

```python
import math
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 5))

#risultati analitici 
call = 14.9758
put = 5.4595

x = np.linspace(0,10**6,100)
C = call*x**0
P = put*x**0

Call, Cerr, Put, Perr= np.loadtxt('./Files/discr.out', delimiter = '  ', unpack = 'True')

plt.subplot(121)
plt.plot(x, C, color='red', label='risultato analitico')  #grafico risultato analitico
plt.errorbar(x, Call, yerr=Cerr, label='simulazione diretta') #grafico simulazione
plt.title("Call-option $C[S(T),T]$")
plt.xlabel('#stime del prezzo')
plt.ylabel('Profitto')
plt.legend()
plt.grid(True)

plt.subplot(122)
plt.plot(x, P, color='red', label='risultato analitico')  #grafico risultato analitico
plt.errorbar(x, Put, yerr=Perr, label='simulazione diretta') #grafico simulazione
plt.title("Put-option $P[S(T),T]$")
plt.xlabel('#stime del prezzo')
plt.ylabel('Profitto')
plt.legend()
plt.grid(True)

plt.show()
```

In entrambe le simulazioni, diretta e discretizzata, ho utilizzato i seguenti valori per i parametri riportati nelle equazioni:

- prezzo degli asset a $t=0$: $S(0)=100$
- istante di tempo finale: $T=1$
- prezzo di mercato: $K=100$
- interesse privo di rischio: $r=0.1$
- volatilità: $\sigma=0.25$
